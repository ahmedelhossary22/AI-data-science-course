Write a program that:
Asks the user to enter a string
Prints the first character of the string
Prints the last character of the string

```python
string=input('Enter a string : ')
print(f'The first character of the string is {string[0]}')
print(f'The last character of the string is {string[-1]}')
```

String Methods
Write a program that:
Asks the user to enter a text
Prints the text in uppercase
Prints the text in lowercase
Prints the number of characters in the text

```python
text=input('Enter a text : ')
print(f'the uppercase of the text is : {text.upper()}')
print(f'the lowercase of the text is : {text.lower()}')
print(f'the number of characters in the text is {len(text)}')
```
