# **Task 1: Compare Two Numbers:**
* Take two numbers as input from the user.

* Use an if condition to print the larger number.

* If the numbers are equal, print "Both are equal".

Example:

Input: 10, 20

Output: 20 is greater

```python
num1=int(input('please enter number 1'))
num2=int(input('please enter number 2'))
if num1 == num2 :
 print("Both are equal")
elif num1 > num2:
 print("num1 is greater")
else:
 print("num2 is greater")

```

```python
num1=int(input('please enter number 1'))
num2=int(input('please enter number 2'))
def greater(num1,num2):
    match num1 == num2:
        case True:
            print("Both are equal")
        case False:
         match num1 > num2:
            case True:
                    print("num1 is greater")
            case False:
                    print("num2 is greater")
greater(num1,num2)                    
```

# **Task 2: Vowel or Consonant**

Ask the user to enter a single letter.

Use an if condition to check if it is a vowel (a, e, i, o, u).

Print "Vowel" if it is a vowel, otherwise print "Consonant".

Example:

Input: e

Output: Vowel


```python
letter = input("enter a single letter ").lower()
if letter=='a' or letter=='e' or letter=='i' or letter=='o' or letter=='u':
    print("vowel")
else:
    print("consonant")
```

```python
letter = input("enter a single letter ").lower()
match letter:
    case 'a' | 'e' | 'i' | 'o' | 'u':
        print("vowel")
    case _:
        print("consonant")
```

# **Task 3: Password Check**

Set a fixed password in the program, e.g., "Open123".

Ask the user to enter a password.

Use an if condition to check if it matches the fixed password.

Print "Access Granted" if correct, otherwise "Access Denied".

Example:

Input: Open123

Output: Access Granted


```python

password = input("Enter your password: ")
match password:
    case "Open123":
        print("Access Granted")
    case _:
        print("Access Denied")
```

```python
fixed_password = "Open123"
password = input("Enter your password: ")
if password == fixed_password:
    print("Access Granted")
else:
    print("Access Denied")
```

# **Task 4: Password Check : Even or Odd**

Write a program that:

Asks the user to enter a number.

If the number is even, print "Even Number".

Otherwise, print "Odd Number".

```python
number=int(input('please enter a number'))
if number%2==0:
    print("number is even")
else:
    print("number is odd")
```

```python
number=int(input('please enter a number'))
match number%2:
    case 0:
        print("number is even")
    case _:
        print("number is odd")
```

# **Task 5: Grade Evaluation**

Write a program that:

Asks the user to enter their score.

Prints:

"Excellent" if the score is 90 or above.

"Very Good" if the score is between 75 and 89.

"Good" if the score is between 60 and 74.

"Fail" if the score is less than 60.

```python
score = int(input('please enter a score: '))

if score >= 90:
    print("excellent")
elif 75 <= score <= 89:
    print("very good")
elif 60 <= score <= 74:
    print("good")
else:
    print("fail")
```

```python
score = int(input('please enter a score: '))
def scores(score):
    match score >= 90:
        case True:
            print("excellent")
        case False:
            match 75 <= score <= 89:
                case True:
                    print("very good")
                case False:
                    match 60 <= score <= 74:
                        case True:
                            print("good")
                        case False:
                            print("fail")
                            
scores(score)
```

# **Task 6: Number Comparison**

Write a program that:

Asks the user to enter a number.

Prints:

"Zero" if the number is 0.

"Positive" if the number is greater than 0.

"Negative" if the number is less than 0.

```python
number=int(input('please enter a number : '))
if number > 0:
    print('positive')
elif number<0:
    print('negative')    
else:
    print('Zero')    
```

```python
number = int(input('please enter a number: '))

match number > 0:
    case True:
        print('positive')
    case False:
        match number < 0:
            case True:
                print('negative')
            case False:
                print('Zero')
```
