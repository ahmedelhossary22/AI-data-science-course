````python
array=[1,2,9,8,5,3]
temp=0
print(f"print the array {array} before sorting")
for i in range(len(array)):
    for j in range(len(array)-1):
        if (array[j]>array[j+1]):
            temp=array[j]
            array[j]=array[j+1]
            array[j+1]=temp
        

print(f"print the array {array} after sorting")
````

Q6.
Write a program that takes a month number (1–12) and prints(use if and match):

"Winter"

"Spring"

"Summer"

"Autumn"

````python
seasons=int(input('please enter the number of seasons: '))
if 3<=seasons<6 :
    print('spring') 
elif 6<=seasons<9:
    print('summer')
elif 9<=seasons<12:
    print('autumn')
else:
    print('winter')    
````

````python
seasons=int(input('please enter the number of seasons: '))
match seasons:
    case 3 | 4 | 5:
        print('spring')
    case 6 | 7 | 8:
        print('summer')
    case 9 | 10 | 11:
        print('autumn')
    case 12 | 1 | 2:
        print('winter')
    case _:
        print('invalid number')
````

Take a character from the user(use if and match).
Check if it is a vowel (a, e, i, o, u) or not.

````python
letter = input("enter a single letter ").lower()
if letter=='a' or letter=='e' or letter=='i' or letter=='o' or letter=='u':
    print("vowel")
else:
    print("consonant")
````

````python
letter = input("enter a single letter ").lower()
match letter:
    case 'a' | 'e' | 'i' | 'o' | 'u':
        print("vowel")
    case _:
        print("consonant")
````

Example
Input


````python
N = int(input('enter the num of row to make the pyramid: '))
for i in range(N):
    for j in range(N - i - 1):
        print(' ', end='')
    for k in range(2*i + 1):
        print('*', end='')
    print()
for i in range(N):
    for j in range(i):
        print(' ', end='')
    for k in range(2*(N-i) - 1):
        print('*', end='')
    print()
````
